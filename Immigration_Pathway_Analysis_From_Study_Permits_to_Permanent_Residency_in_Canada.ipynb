```python
df_pr = df_pr.rename(columns={"Unnamed: 3": "Program"})
df_pr.head()
```

```python
df_pr_clean = df_pr[["Program", "Jan", "Feb", "Mar", "Apr", "May", "Jun"]].copy()
df_pr_clean.head(10)
```

```python
df_pr_clean = df_pr_clean[df_pr_clean["Program"].notna()].copy()

df_pr_clean = df_pr_clean[
    ~df_pr_clean["Program"].astype(str).str.contains(
        "Notes|Source|For further information|Total",
        na=False
    )
].copy()

df_pr_clean = df_pr_clean.dropna(
    subset=["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    how="all"
)

df_pr_clean.head(10)
```

```python
df_pr_long = df_pr_clean.melt(
    id_vars=["Program"],
    var_name="Month",
    value_name="PR_Count"
)

df_pr_long["PR_Count"] = pd.to_numeric(df_pr_long["PR_Count"], errors="coerce")

df_pr_long.head(10)
```

```python
df_program = df_pr_long.groupby("Program")["PR_Count"].sum().reset_index()
df_program = df_program[df_program["PR_Count"] > 0]

df_program
```

```python
plt.figure()
plt.bar(df_program["Program"], df_program["PR_Count"])
plt.xticks(rotation=60)
plt.title("PR by Immigration Program")
plt.xlabel("Program")
plt.ylabel("PR Count")
plt.show()
```

## Key Insight: PR Transition Patterns

The analysis shows that the majority of permanent residency transitions occur through the Skilled Worker program, followed by Sponsored Spouse or Partner and Canadian Experience.

This indicates that economic immigration pathways dominate PR outcomes, while smaller programs contribute relatively little to overall PR transitions.

When compared with study permit distribution, this suggests that international students may ultimately rely on a limited number of dominant immigration pathways when transitioning to permanent residency.

## Final Conclusion

This project explored the broader pathway from study permits to permanent residency in Canada using two real-world datasets.

The findings show that:
- Study permits are concentrated in a few key provinces
- Permanent residency transitions are dominated by a small number of immigration programs

Although the datasets do not track individual-level transitions, the analysis highlights how both geographic distribution and immigration program structure influence the pathway to permanent residency.
